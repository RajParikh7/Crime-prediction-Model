```python
import pandas as pd
from datetime import datetime
```

```python
df=pd.read_csv('crime_2011_9_chicago.csv')
```

```python
df.columns
```

```python
df=df.rename_axis({'Date':'DateTime'},axis="columns")
```

```python
df
```

```python
df["DateTime"] = df["DateTime"].apply(lambda x:datetime.strptime(x,"%m/%d/%Y %I:%M:%S %p"))
```

```python
df['Date']=[d.date() for d in df['DateTime']]
df['Time']=[d.time() for d in df['DateTime']]
```

```python
df['Hour']=df['DateTime'].dt.hour
```

```python
df['DateTime'].rename_axis('Date',axis=1)
```

```python
df=df.dropna(axis=0)
```

```python
df.to_csv(r'crime_2011_9_Chicago.csv')
```

```python
df1=df.groupby(['Date','Hour','Community Area','Primary Type'],as_index=False)['DateTime'].count()
```

```python
for i in range(len(df1)):
    df1.loc[i,'Hour']=chr(ord('A')+df1.loc[i,'Hour'])
```
